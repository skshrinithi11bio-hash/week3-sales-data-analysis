```python
import pandas as pd

# Creating a sample sales dataset
data = {
    "Product": ["Laptop", "Phone", "Tablet", "Laptop", "Phone", "Tablet", "Laptop"],
    "Quantity": [2, 5, 3, 1, None, 4, 2],
    "Price": [60000, 30000, 20000, 60000, 30000, None, 60000]
}

df = pd.DataFrame(data)

# Save to CSV
df.to_csv("sales_data.csv", index=False)

df
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("sales_data.csv")

# Handle missing values (NO inplace warnings)
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].mean())
df["Price"] = df["Price"].fillna(df["Price"].mean())

# Calculate total sales
df["Total_Sales"] = df["Quantity"] * df["Price"]

# Metrics
total_revenue = df["Total_Sales"].sum()
average_sales = df["Total_Sales"].mean()
best_selling_product = df.groupby("Product")["Total_Sales"].sum().idxmax()

# -------- CLEAN FINAL OUTPUT ONLY --------
print("ðŸ“Š SALES ANALYSIS REPORT")
print("------------------------")
print(f"Total Revenue       : â‚¹{total_revenue:,.2f}")
print(f"Average Sales Value : â‚¹{average_sales:,.2f}")
print(f"Best Selling Product: {best_selling_product}")
```
